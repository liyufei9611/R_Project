```python
import requests
import os
import time
import re
from bs4 import BeautifulSoup 

headers={'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}

id=200
for year in date_dic:
    for date in date_dic[year]:
        for i in range(1,9):
            url='https://www.mjc.mo/race/info/dividend.php?selectYr={}&race_date={}&location=MC&race_no={}'.format(year,date,i)
            #print(url)
            response=requests.get(url=url,headers=headers)
            if len(response.text)<42000:
                continue
            
            soup=BeautifulSoup(response.text, "html.parser")  
            info=soup.find_all(attrs={'align':"left",'valign':"top"})[0].string.split('|')
            with open('./mjc_crawl.txt', 'a') as f:
                f.write('I, '+str(id)+','+str(i)+','+",".join(info) + os.linesep)
                
            table=soup.find_all(href=re.compile('www.mjc.mo/race/info/hrs_info'))[-1].parent.parent.parent
            
            for record in table.find_all('tr')[1:]:
                il=[]
                for element in record.find_all('td')[:-1]:
                    il.append(element.text)
                with open('./mjc_crawl.txt', 'a') as f:
                    f.write(str(id)+','+",".join(il) + os.linesep)
            id=id+1
                    
        time.sleep(3)         
        print('finished: '+str(year)+' '+date+' '+str(id))
        
            
```
